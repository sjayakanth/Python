```python
import pandas as pd
```

```python
data = pd.read_csv('.././csv/OnlineRetail.csv',encoding='latin1')
df = pd.DataFrame(data=data)
```

```python
df.head(5)
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.info()
```

```python
#what is the unique count of the invoice numbers:
len(pd.unique(df['InvoiceNo'])) #25900
```

```python
#How many people bought “PARTY BUNTING”.:*
df[df['Description']=='PARTY BUNTING'].groupby('CustomerID').sum()
```

```python
#how many unique people are from “UNITED KINGDOM”
df_country = df[['CustomerID','Country']].groupby('Country').agg({'CustomerID':'count'}).reset_index()
df_country[df_country['Country']=='United Kingdom']
```

```python
#Which customer(CustomerID) has done highest transaction in the dataset.
df['price'] = df['Quantity'] * df ['UnitPrice']
```

```python
df.head()
```

```python
df.sort_values(by='price',ascending=False)
```

```python
#What is the average amount of transactions
df['price'].mean()
```

```python
#What is the frequently bought item in the dataset*
df_desc = df[['Description']].groupby('Description').agg({'Description':'count'})
df_desc.columns = ['count']
```

```python
df_desc.reset_index()
df_desc.sort_values('count',ascending=False)
```
